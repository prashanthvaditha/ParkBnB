```python
#example of python connecting to MySQL server and databases
#
import mysql.connector
#
from mysql.connector import Error
#

# import json
# with open('password.json', 'r') as f:
#     data = json.load(f)
# password = data['my_password']


try:
    connection = mysql.connector.connect(host='localhost',
                                         database='parkbnb_final',
                                         user='root',
                                         password=f'Prashanth@123',
                                         auth_plugin = 'mysql_native_password')
    if connection.is_connected():
        db_Info = connection.get_server_info()
        print("Connected to MySQL Server version ", db_Info)
        cursor = connection.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("Your connected to database: ", record)
#
        sql_select_Query = "select * from user limit 10"
        cursor = connection.cursor()
        cursor.execute(sql_select_Query)
        records = cursor.fetchall()
        print("Countries with english language:\n")
        for row in records:
            print(row,"\n")
#
except Error as e:
    print("Error while connecting to MySQL", e)
finally:
    try:
        if (connection.is_connected()):
            cursor.close()
            connection.close()
            print("MySQL connection is closed")
    except:
        print("Error")


#you should see the following output
#'''Connected to MySQL Server version  8.0.17
#Your connected to database:  ('classicmodels',)
#True
#MySQL connection is closed'''
#


```
